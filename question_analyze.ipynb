```python
#將df的影像欄位都轉成img address
image_columns = ['題幹圖一','題幹圖二','圖A','圖B','圖C','圖D']
notna_filter = df[image_columns[0]].notna()
# 'image/01生命世界/image1.jpeg'
'image/' +  df[notna_filter]['主題'] + '/image'  + df[notna_filter][image_columns[0]].astype(int).apply(str) + '.jpeg'
```

```python
#增加欄位，存放題幹和選項的圖片路徑
image_src_columns = ['題幹圖一','題幹圖二','圖A','圖B','圖C','圖D']
image_dst_columns = ['stem_img_src','a_img_src','b_img_src','c_img_src','d_img_src']

for column_name in image_dst_columns:
    df[column_name] = ''

    
def cal_stem_img(row):
    if pd.notna(row['題幹圖一']):
        row['stem_img'] = 'image/' +  row['主題'] + '/image' + str(int(row['題幹圖一'])) + '.jpeg'
    else:
        row['stem_img'] = ''
 
    
        
    return row


df3 = df.apply(cal_stem_img, axis=1)  # 記得: 1.要把新的df指派給原本的df才會成功更新。 2. axis要設定為1，才會以row為單位跑回圈。
df3
```

```python
#增加欄位，存放題幹和選項的圖片路徑
image_source_columns = ['題幹圖一','題幹圖二','圖A','圖B','圖C','圖D']
image_dst_columns = ['stem_img1_src','stem_img2_src','a_img_src','b_img_src','c_img_src','d_img_src']

for column_name in image_dst_columns:
    df[column_name] = ''


    
def cal_img_to_address(row):
    for i in range( len(image_source_columns) ):
        if pd.notna(row[ image_source_columns[i] ]):
            row[ image_dst_columns[i] ] = 'image/' +  row['主題'] + '/image' + str(int(row[ image_source_columns[i] ])) + '.jpeg'
        else:
            row[ image_dst_columns[i] ] = ''
   
    return row    


df3 = df.apply(cal_img_to_address, axis=1)  # 記得: 1.要把新的df指派給原本的df才會成功更新。 2. axis要設定為1，才會以row為單位跑回圈。
df3
```

```python
HTML(df3.to_html(escape=False ,formatters=dict( stem_img1_src = path_to_image_html,
                                               stem_img2_src = path_to_image_html,
                                               a_img_src = path_to_image_html,
                                               b_img_src = path_to_image_html,
                                               c_img_src = path_to_image_html,
                                               d_img_src = path_to_image_html,
                                              )))
```

```python
df3.to_html('df3.html', escape = False, formatters=dict( stem_img1_src = path_to_image_html,
                                               stem_img2_src = path_to_image_html,
                                               a_img_src = path_to_image_html,
                                               b_img_src = path_to_image_html,
                                               c_img_src = path_to_image_html,
                                               d_img_src = path_to_image_html,
                                              ))
```
