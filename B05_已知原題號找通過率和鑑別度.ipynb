```python
passing_df = pd.read_csv('合併後的通過率與鑑別度.csv')  
print(passing_df)  
```

```python
df = pd.read_csv('database.csv', dtype=str, encoding = 'utf-8')

df
```

```python
year = 103
number = 3
ref1 = passing_df['年度'] == year
ref2 = passing_df['科目'] == '自然'
ref3 = passing_df['題號'] == number

dis_index = passing_df[ref1 & ref2 & ref3]['鑑別度'].values.item(0)
pass_index = passing_df[ref1 & ref2 & ref3]['通過率'].values.item(0)
print(dis_index, pass_index)
```

```python
year = 103


df_filter1 = df['出處'].str.contains(str(year))


df_year = df[df_filter1]

for num in  df_year['原題號']:
    
    ref1 = passing_df['年度'] == year
    ref2 = passing_df['科目'] == '自然'
    ref3 = passing_df['題號'] == int(num)
    
    dis_index  = passing_df[ref1 & ref2 & ref3]['鑑別度'].values.item(0)
    pass_index = passing_df[ref1 & ref2 & ref3]['通過率'].values.item(0)

    print(str(year), num , dis_index, pass_index)
    
    df_filter2 = df['原題號'] == num

    
    df.loc[ df_filter1 & df_filter2 , '鑑別度'] = dis_index
    df.loc[ df_filter1 & df_filter2 , '通過率'] = pass_index
    
```
