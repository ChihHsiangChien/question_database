```python
inquiry_df = pd.read_csv('會考題目/探究實作題目.csv')  
print(inquiry_df)  
```

```python
def drawImage(year, quesiton_num, item_pass,item_discri):
    subject = '自然'
    x = passing_df[passing_df['科目'] == subject]['通過率']
    y = passing_df[passing_df['科目'] == subject]['鑑別度']

    fig, ax = plt.subplots()

    #畫出所有的題目
    ax.plot(x, y, marker='o', linestyle='', ms = 7,color = ((0.1, 0.1, 0.1, 0.15)))

    #畫出單題
    #item_pass = 0.2
    #item_discri = 0.4
    ax.plot(item_pass, item_discri, marker='o', linestyle='', ms=5,color =((1, 0, 0, 0.9)))
    ax.text(0.1, 0.7, str(year) + '年' , fontsize = 5, fontproperties=myfont)
    ax.text(0.1, 0.65, str(quesiton_num) + '題' , fontsize = 20, fontproperties=myfont)
    ax.text(0.1, 0.60, '通過率:' + str(item_pass) , fontsize = 20, fontproperties=myfont)
    ax.text(0.1, 0.55, '鑑別度:' + str(item_discri) , fontsize = 20, fontproperties=myfont)
    
    


    ax.set_title(subject,  fontsize = 20, fontproperties = myfont)
    ax.set_xlabel('通過率' , fontsize = 20, fontproperties = myfont)
    ax.set_ylabel('鑑別度' , fontsize = 20, fontproperties = myfont)
    ax.grid()
    ax.set_xticks((0, 0.5, 1))
    ax.set_yticks((0, 0.2, 0.5, 1))

    ax.set_xlim([0, 1])
    ax.set_ylim([0, 0.8])



    plt.savefig('探究實作題目/'+ str(year) + '_' + str(quesiton_num) +'.png', dpi = 150)
    plt.show()
```

```python
for i in range(inquiry_df.shape[0]):
#for i in range(1):    
    year = inquiry_df['年度'].iloc[i]
    quesiton_num = inquiry_df['題號'].iloc[i]
    if year <= 109:

        filter_1 = passing_df['科目'] == subject
        filter_2 = passing_df['年度'] == year
        filter_3 = passing_df['題號'] == quesiton_num

        item_pass = passing_df[filter_1 & filter_2 & filter_3]['通過率'].values[0]
        item_discri = passing_df[filter_1 & filter_2 & filter_3]['鑑別度'].values[0]
        print(year, quesiton_num, item_pass,item_discri)
        drawImage(year, quesiton_num, item_pass,item_discri)
        #print(item_discri)
```
