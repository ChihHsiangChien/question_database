```python
for year in range(104,110):
    #year = 104


    filter1 = question_df['年度'] == year
    filter2 = question_df['科目'] == '自然'
    
    df_nature = question_df[ filter1 & filter2 ]


    df_year = df[df['出處'].str.contains(str(year))]['題幹']
    for sentence in df_year:
        sentence_seg = sentence[-20:-1]    
        
        dd = df_nature['題目'].str.contains(sentence_seg)
        #print(sentence_seg, str(year), df_nature[dd]['題號'])
        ee = df_nature[dd]['題號']
        for number in ee:
            
            
            ref1 = passing_df['年度'] == year
            ref2 = passing_df['科目'] == '自然'
            ref3 = passing_df['題號'] == number

            dis_index = passing_df[ref1 & ref2 & ref3]['鑑別度'].values.item(0)
            pass_index = passing_df[ref1 & ref2 & ref3]['通過率'].values.item(0)
            
            print(sentence_seg, str(year), number , dis_index, pass_index)
            

            
        #print(df_nature[dd]['題目'])
```
