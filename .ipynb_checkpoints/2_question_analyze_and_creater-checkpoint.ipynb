```python
import pandas as pd
from IPython.display import Image, HTML
import math
import numpy as np

pd.set_option('display.width', 200, 'display.max_rows', 1000)


#讀取檔案
#df = pd.read_excel("題庫.xlsx", dtype=str,encoding="utf-8")
df = pd.read_csv('database.csv', dtype=str, encoding = 'utf-8')


#將所有欄位轉成string
df = df.astype(str)
df = df.replace('nan', '')
#df.head(5)
```

```python
#增加欄位，存放題幹和選項的圖片路徑
image_source_columns = ['題組圖1','題組圖2','題幹圖1','題幹圖2','圖A','圖B','圖C','圖D']
image_dst_columns = ['group_img1_src','group_img2_src','stem_img1_src','stem_img2_src','a_img_src','b_img_src','c_img_src','d_img_src']

for column_name in image_dst_columns:
    df[column_name] = ''

#處理NaN轉成""，欄位格式還是float，無法取len
    
def cal_img_to_address(row):
    for i in range( len(image_source_columns) ):
        if len(row[ image_source_columns[i] ]) > 0 :
            row[ image_dst_columns[i] ] = 'image/' +  str(row[ image_source_columns[i]]) + '.jpeg'
        else:
            row[ image_dst_columns[i] ] = ''
   
    return row    

df = df.apply(cal_img_to_address, axis=1)  # axis設定為1，以row為單位跑loop。
#df.head(5)
```

# 匯出

```python
#用multiindex的 df輸出題庫

#將dataframe 變成 multindex
data = df.set_index(['章', '節', '概念'])

#將 multiindex的df排序
data = data.sort_index()

#檢查是否已經排序
data.index.is_lexsorted()


#對題幹的圖和選項的圖分別處理
def add_img_src(column_name):
    if len(group_data[ column_name ].values[i] ) > 0:
        global question_txt
        question_txt = question_txt + '<img src = "'
        question_txt = question_txt + group_data[ column_name ].values[i]
        if column_name == 'stem_img1_src' or column_name == 'stem_img2_src':
            question_txt = question_txt + '" height="100" align="right">'
        else:
            question_txt = question_txt + '" height="100"  >'

with open('題庫.html', 'w', encoding='utf-8' ) as myFile:
    myFile.write('<html>')
    myFile.write('<head>')
    myFile.write('<title>')
    myFile.write('生物會考與基冊考題整理')    
    myFile.write('</title>')
    myFile.write('</head>')
        
    myFile.write('<body>')

    
    for k in range(len(data.index.unique())):
        sector =  data.index.unique()[k]

        question_txt = '<hr>'        
        question_txt = question_txt + '<H1>'
        question_txt = question_txt + sector[0] + '-' + sector[1] + '：' + sector[2] 
        question_txt = question_txt + '</H1>'
        
        
        group_data = data.loc[sector[0] , sector[1] , sector[2]].sort_values(by=['出處'])
        question_txt = question_txt + '共' + str(len(group_data )) + '題'
        
        question_txt = question_txt + '<hr>'

        
        idx = 1  #題號
        
        myFile.write(question_txt)
        
        for i in range ( len(group_data ) ):

            question_txt = '<p>'

            add_img_src('stem_img1_src')
            add_img_src('stem_img2_src')
            
            question_txt = question_txt + str(idx) + '.(&emsp;&emsp;)'
            question_txt = question_txt + group_data['題幹'].values[i]

            question_txt = question_txt + '【'        
            question_txt = question_txt + group_data['出處'].values[i]
            question_txt = question_txt + '】'

            question_txt = question_txt + '<br>'
            question_txt = question_txt + '&emsp;'
            question_txt = question_txt + '(A)'
            question_txt = question_txt + group_data['選項A'].values[i]
            add_img_src('a_img_src')
            question_txt = question_txt + '&emsp;'

            question_txt = question_txt + '<br>'
            question_txt = question_txt + '&emsp;'
            question_txt = question_txt + '(B)'
            question_txt = question_txt + group_data['選項B'].values[i]
            add_img_src('b_img_src')
            question_txt = question_txt + '&emsp;'

            question_txt = question_txt + '<br>'
            question_txt = question_txt + '&emsp;'
            question_txt = question_txt + '(C)'    
            question_txt = question_txt + group_data['選項C'].values[i]
            add_img_src('c_img_src')
            question_txt = question_txt + '&emsp;'

            question_txt = question_txt + '<br>'
            question_txt = question_txt + '&emsp;'
            question_txt = question_txt + '(D)'    
            question_txt = question_txt + group_data['選項D'].values[i]
            add_img_src('d_img_src')
            question_txt = question_txt + '&emsp;'

            question_txt = question_txt + '<hr>'
            
            myFile.write(question_txt)
            
            idx = idx + 1

    
    
    myFile.write('</body>')
   
    myFile.write('</html>')
```
