```python
#增加欄位，存放題幹和選項的圖片路徑
image_source_columns = ['題幹圖一','題幹圖二','圖A','圖B','圖C','圖D']
image_dst_columns = ['stem_img1_src','stem_img2_src','a_img_src','b_img_src','c_img_src','d_img_src']

for column_name in image_dst_columns:
    df[column_name] = ''

#處理NaN轉成""，欄位格式還是float，無法取len
    
def cal_img_to_address(row):
    for i in range( len(image_source_columns) ):
        if len(row[ image_source_columns[i] ]) > 0 :
            row[ image_dst_columns[i] ] = 'image/' +  row['章'] + '/image' + str(int(row[ image_source_columns[i] ])) + '.jpeg'
        else:
            row[ image_dst_columns[i] ] = ''
   
    return row    


df = df.apply(cal_img_to_address, axis=1)  # 記得: 1.要把新的df指派給原本的df才會成功更新。 2. axis要設定為1，才會以row為單位跑回圈。
df
```

```python
html = df.to_html(escape = False, formatters=dict( stem_img1_src = path_to_image_html,
                                               stem_img2_src = path_to_image_html,
                                               a_img_src = path_to_image_html,
                                               b_img_src = path_to_image_html,
                                               c_img_src = path_to_image_html,
                                               d_img_src = path_to_image_html,
                                              ))
with open('df3.html', 'w', encoding = 'utf-8') as file:
    file.write(html)
```

```python
#輸出題庫的測試
colour = ["red", "red", "green", "yellow"]

with open('mypage.html', 'w') as myFile:
    myFile.write('<html>')
    myFile.write('<body>')
    myFile.write('<table>')

    s = '1234567890'
    for i in range(0, len(s), 60):
        myFile.write('<tr><td>%04d</td>' % (i+1));
    for j, k in enumerate(s[i:i+60]):
        myFile.write('<td><font style="background-color:%s;">%s<font></td>' % (colour[j %len(colour)], k));


    myFile.write('</tr>')
    myFile.write('</table>')
    myFile.write('</body>')
    myFile.write('</html>')
```

```python
def add_img_src(column_name):
    if len(row[ column_name ]) > 0:
        global question_txt
        question_txt = question_txt + '<img src = "'
        question_txt = question_txt + row[ column_name ]
        if column_name == 'stem_img1_src' or column_name == 'stem_img2_src':
            question_txt = question_txt + '" height="100" align="right">'
        else:
            question_txt = question_txt + '" height="100"  >'
```

```python
#輸出題庫
with open('題庫.html', 'w', encoding='utf-8' ) as myFile:
    myFile.write('<html>')
    myFile.write('<head>')
    myFile.write('<title>')
    myFile.write('生物會考與基冊考題整理')    
    myFile.write('</title>')
    myFile.write('</head>')
        
    myFile.write('<body>')

    
    for idx, row in df.iterrows():
        question_txt = '<p>'

        add_img_src('stem_img1_src')
        add_img_src('stem_img2_src')
        
        question_txt = question_txt + row['題幹'] 
        
        question_txt = question_txt + '【'        
        question_txt = question_txt + row['出處']
        question_txt = question_txt + '】'
        
        question_txt = question_txt + '<br>'
        question_txt = question_txt + '&emsp;'
        question_txt = question_txt + '(A)'
        question_txt = question_txt + row['選項A']
        add_img_src('a_img_src')
        question_txt = question_txt + '&emsp;'

        question_txt = question_txt + '<br>'
        question_txt = question_txt + '&emsp;'
        question_txt = question_txt + '(B)'
        question_txt = question_txt + row['選項B']
        add_img_src('b_img_src')
        question_txt = question_txt + '&emsp;'

        question_txt = question_txt + '<br>'
        question_txt = question_txt + '&emsp;'
        question_txt = question_txt + '(C)'    
        question_txt = question_txt + row['選項C']
        add_img_src('c_img_src')
        question_txt = question_txt + '&emsp;'

        question_txt = question_txt + '<br>'
        question_txt = question_txt + '&emsp;'
        question_txt = question_txt + '(D)'    
        question_txt = question_txt + row['選項D']
        add_img_src('d_img_src')
        question_txt = question_txt + '&emsp;'


        
        question_txt = question_txt + '<hr>'
        myFile.write(question_txt)
    
    
    
    myFile.write('</body>')
   
    myFile.write('</html>')
```

```python
#按章節印出題目

sectors = df['節'].groupby(df['章']).value_counts( sort = False)
sectors = sectors.index
for sector in sectors:
    print( '*****' + sector[0] + '*****' +  sector[1] + '*****')
    
    group_data = data.loc[sector[0] , sector[1]]
    for i in range ( len(group_data ) ):
        print( group_data['題幹'].values[i] )
        print( '(A) ' + group_data['選項A'].values[i] )
        print( '(B) ' + group_data['選項B'].values[i] )
        print( '(C) ' + group_data['選項C'].values[i] )
        print( '(D) ' + group_data['選項D'].values[i] )
    print('*************************')
```

```python
#將dataframe 變成 multindex
data = df.set_index(['章', '節', '概念'])

#將 multiindex的df排序
data = data.sort_index()

#檢查是否已經排序
data.index.is_lexsorted()

for i in range(len(data.index.unique())):
    sector =  data.index.unique()[i]
    print('*************************')
    print(sector[0] + ':' + sector[1] + '----' + sector[2] )
    print('*************************')
    group_data = data.loc[sector[0] , sector[1] , sector[2]]
    for i in range ( len(group_data ) ):
        print( group_data['題幹'].values[i] )
        print( '(A) ' + group_data['選項A'].values[i] )
        print( '(B) ' + group_data['選項B'].values[i] )
        print( '(C) ' + group_data['選項C'].values[i] )
        print( '(D) ' + group_data['選項D'].values[i] )
        print( '' )
    print('================================================================')
```
