```python
import pandas as pd
import math
file = "會考通過率和鑑別度.xlsx"

#讀取sheet_name
excel = pd.ExcelFile(file)
print(excel.sheet_names)
sheet_counts = len(excel.sheet_names)
print(sheet_counts)
```

```python
df = pd.read_excel( file, sheet_name = None)
print(df)
```

```python
df.columns.values
```

```python
def getExcelValue(sheet):
    sheetName = excel.sheet_names[sheet]
    df = pd.read_excel( file, sheet_name = sheetName)


    #df的資料長度
    df.shape[0]

    #df的欄數
    df.shape[1]

    #所有欄的名稱
    df.columns.values


    for col in df.columns.values[1:-1]:

        #讀取儲存格的值(以列索引值及欄位標題來定位)
        for row in range(df.shape[0]):
            #val = df.at[0, col] 
            #print(sheetName)
            #print(sheetName[0:3])
            #print(sheetName[3:])

            #print(col)
            #print(row)
            #print(df[col][row])
            
            #if not Nan就print
            if math.isnan(df[col][row]) != True :
                #print(sheetName[0:3],sheetName[3:], col, str(row+1), df[col][row])
                new = pd.DataFrame.from_dict( {"年度": [sheetName[0:3]],
                               "類別":[sheetName[3:]],
                               "科目":[col],
                               "題號":[str(row+1)],
                               "數值":[df[col][row]]})
                global new_df

                new_df = new_df.append(new,ignore_index=True)   # ignore_index=True,表示不按原來的索引，從0開始自動遞增
```

```python
dd = {"年度":[], "類別":[], "科目":[], "題號":[], "數值":[]}
new_df = pd.DataFrame.from_dict(dd)

for i in range(sheet_counts):
    getExcelValue(i)

new_df
```

```python
new_df.to_csv('整理後的通過率與鑑別度.csv', index=False,encoding="utf_8_sig")
```
