```python
import os
from win32com import client as wc
import docx
```

```python
#doc to docx
source_directory = r"C:\Users\user\Downloads\test"
dst_directory = os.getcwd()
word = wc.Dispatch('Word.Application')

for process_file in  os.listdir(source_directory):
    file, extension = os.path.splitext(process_file)
    
    source_file_fullpath = os.path.join(source_directory, process_file)
    dest_file_path = file + '.docx'
    dest_file_fullpath = os.path.join(dst_directory, dest_file_path)
    print(dest_file_fullpath)
    doc = word.Documents.Open(source_file_fullpath)        # 目標路徑下的檔案
    doc.SaveAs(dest_file_fullpath, 12, False, "", True, "", False, False, False, False)  # 轉化後路徑下的檔案    
    doc.Close()
    
word.Quit()
    
```
