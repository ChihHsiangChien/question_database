```python
#增加欄位，存放題幹和選項的圖片路徑
image_source_columns = ['題幹圖一','題幹圖二','圖A','圖B','圖C','圖D']
image_dst_columns = ['stem_img1_src','stem_img2_src','a_img_src','b_img_src','c_img_src','d_img_src']

for column_name in image_dst_columns:
    df[column_name] = ''

#處理NaN轉成""，欄位格式還是float，無法取len
    
def cal_img_to_address(row):
    for i in range( len(image_source_columns) ):
        if len(row[ image_source_columns[i] ]) > 0 :
            row[ image_dst_columns[i] ] = 'image/' +  row['主題'] + '/image' + str(int(row[ image_source_columns[i] ])) + '.jpeg'
        else:
            row[ image_dst_columns[i] ] = ''
   
    return row    


df = df.apply(cal_img_to_address, axis=1)  # 記得: 1.要把新的df指派給原本的df才會成功更新。 2. axis要設定為1，才會以row為單位跑回圈。
df
```

```python
def add_img_src(column_name):
    if len(row[ column_name ]) > 0:
        global question_txt
        question_txt = question_txt + '<img src = "'
        question_txt = question_txt + row[ column_name ]
        if column_name == 'stem_img1_src' or column_name == 'stem_img2_src':
            question_txt = question_txt + '" height="100" align="right">'
        else:
            question_txt = question_txt + '" height="50"  >'
```

```python
with open('題庫.html', 'w') as myFile:
    myFile.write('<html>')
    myFile.write('<head>')
    myFile.write('<title>')
    myFile.write('生物會考與基冊考題整理')    
    myFile.write('</title>')
    myFile.write('</head>')
        
    myFile.write('<body>')

    
    for idx, row in df.iterrows():
        question_txt = '<p>'

        add_img_src('stem_img1_src')
        add_img_src('stem_img2_src')
        
        question_txt = question_txt + row['題幹']          
        
        question_txt = question_txt + '<br>'
        question_txt = question_txt + '(A)'
        question_txt = question_txt + row['選項A']
        add_img_src('a_img_src')
        question_txt = question_txt + '&emsp;'

        question_txt = question_txt + '(B)'
        question_txt = question_txt + row['選項B']
        add_img_src('b_img_src')
        question_txt = question_txt + '&emsp;'

        question_txt = question_txt + '(C)'    
        question_txt = question_txt + row['選項C']
        add_img_src('c_img_src')
        question_txt = question_txt + '&emsp;'

        question_txt = question_txt + '(D)'    
        question_txt = question_txt + row['選項D']
        add_img_src('d_img_src')
        question_txt = question_txt + '&emsp;'

        question_txt = question_txt + '【'        
        question_txt = question_txt + row['出處']
        question_txt = question_txt + '】'
        question_txt = question_txt + '<hr>'
        myFile.write(question_txt)
    
    
    
    myFile.write('</body>')
   
    myFile.write('</html>')
```
